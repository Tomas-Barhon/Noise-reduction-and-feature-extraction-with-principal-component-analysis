```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
# Set plot style
plt.style.use('seaborn-v0_8-pastel')

# Generate data
np.random.seed(0)
mean = [0, 0]
cov = [[3, 1], [1, 1]]  # diagonal covariance
data = np.random.multivariate_normal(mean, cov, 500)

# Perform PCA
pca = PCA(n_components=2)
pca.fit(data)
components = pca.components_

# Plot data
plt.figure(figsize=(10, 6))
plt.scatter(data[:, 0], data[:, 1], alpha=0.6, edgecolors='w', s=50, label='Data Points')


plt.quiver(0, 0, components[0, 0], components[0, 1], color='r', scale=3, label='Max Variance Hyperplane')


plt.quiver(0, 0, components[1, 0], components[1, 1], color='b', scale=3, label='Lower Variance Hyperplane')
plt.grid(False)
plt.legend()

plt.xlabel('X')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.ylabel('Y')
plt.savefig('./../text/Figures/pca_plot.pdf', format='pdf', dpi=300)
plt.show()
```
